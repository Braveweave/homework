```python
from math import log


log2 = lambda x: log(x, 2)


def entropy(df, result_row):
    '''Calculate entropy from data frame.'''
    e = 0.0
    total = float(len(df))
    
    for name, group in df.groupby(result_row):
        pg = len(group) / total
        e = e + log2(pg) * pg
        
    return -e


def gain(df, attribute_row, result_row):
    '''Calculate gain entropy from data frame.'''
    e = 0.0
    total = float(len(df))
    
    for name, group in df.groupby(attribute_row):
        pg = len(group) / total
        entropy_group = entropy(group, result_row)
        
        e = e + pg * entropy_group
        
    return e


INSTANCE_ATTRIBUTE = 'instance'


def tree(df, result_row):
    
    def divide(df, attribute_row):
        decision = {}
        
        for name, group in df.groupby(attribute_row):
            field = (attribute_row, name)
            epy = entropy(group, result_row)
            if abs(epy) == 0.0:
                decision[field] = group[result_row].get_values()[0]
            else:
                decision[field] = tree(group, result_row)
                
        return decision
    
    min_gain, min_attribute = float('inf'), None
    for attribute in df.columns:
        if attribute == INSTANCE_ATTRIBUTE or attribute == result_row:
            continue
        attribute_gain = gain(df, attribute, result_row)
        if attribute_gain < min_gain:
            min_gain = attribute_gain
            min_attribute = attribute
            
    return divide(df, min_attribute)


def explain(tree):
    
    def visit(tree):
        if isinstance(tree, str):
            print(tree)
            return
        for cond, action in tree.items():
            print('{0} = {1}'.format(*cond))
            visit(action)
    
    visit(tree)
```
